### Step 3: Exploratory Analysis and Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

### *Question 1: What is the correlation between all columns? What does the correlation tells us about loan and income?*

```python
warnings.filterwarnings("ignore")
corr_matrix = df.corr()
plt.figure(figsize=(10,5))
sns.heatmap(corr_matrix, annot=True, fmt='.2f')
plt.title('Correlation Matrix to show how the columns are related', fontsize = 15)
plt.show()
```

**Interpretation:**

The graph above states that there is a strong correlation between some features and a weak correlation between the others. The closer to 1, the stronger positive correlation and the closer to -1, the stronger the negative correlation. It can be seen that income annual and loan amount has a strong positive correlation.






**Answer 1:**
It can be interpreted from the graph that as the annual income increases the loan amount tends to increase as well.

### *Question 2: How strongly does income impact the decision of loan approval?*

```python
# Boxplot for income based on loan status
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['loan_status'], y=df['income_annum'])
plt.title('Annual Income Distribution by Loan Status')
plt.xlabel('Loan Status')
plt.ylabel('Annual Income')
plt.show()
```

**Interpretation**

The above graph suggests that median income of rejected loans is slightly above than approved income. It can also be seen that both categories overkao the band of income. Which means that while one indivisual with a certain income was rejected, the other might have been approved. This suggests that more factors are taken into consideratation when it comes to loan approval than just the income of an individual.

**Answer 2:**

While income is an important consideration, it is not the only factor that plays a role in the decision of approval or rejection.

### *Question 3: Does number of dependants influence an individuals loan status?*

```python
# Cross-tabulation
dependents_approval = pd.crosstab(df['no_of_dependents'], df['loan_status'])

# Plotting
dependents_approval.plot(kind='bar', stacked=True)
plt.title('Number of Dependents vs Loan Approval Status')
plt.ylabel('Count')
plt.show()
```

**Interpretation:**

The above graph shows from dependants 0-2, the ratio of approved is higher than of those that were rejected. As the number of dependants increase, the ratio changes slightly. Individuals 3, 4, and 5 dependents have a greater or equal number of rejections compared to approvals.

**Answer 3:**

There may be association between loan status approval with having few or none dependants. This can be seen from a perspective of bank how an individuals income is distributed/spent. The two factors of income and dependants can be assesed together to get a stronger decision of approval or rejection.

### *Question 4: Is there a minimum amount of CIBIL score needed for approved loan status? How big of a role does CIBBIL score plays in loan status?*

```python
# Scatter plot for CIBIL score and loan amount
plt.figure(figsize=(8, 5))
sns.scatterplot(x='cibil_score', y='loan_amount', data=df, hue='loan_status')
plt.title('Loan Amount vs CIBIL Score by Loan Status')
plt.show()
```

**Interpretation:**

From the above scatterplot, it can be seen that there is a spread of approved and rejected all across the scale. The majority of appoved lies between 600-900 of scale.

**Answer 4:**

A 600+ score of CIBIL score is needed to make an individuals application strong that will lead to approval of the loan. Since there are approved and Rejected between both areas, it shows that while having a strong store does increase chances of approval, it is not the only factor that is considered in the final decision of loan approval or rejection. The people who might have been approved with a lower CIBIL score might have other factors such as income, or assets that play a role in the decision.

Encoding data preprocessing labels

```python
# Import the library
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# load the dataset
df = pd.read_csv("loan_approval_dataset (1).csv")

# Perform label encoding for categorical variables
label_encoder = LabelEncoder()
categorical_columns = df.select_dtypes(include=['object']).columns
for col in categorical_columns:
    df[col] = label_encoder.fit_transform(df[col].astype(str).str.strip())

# Data segmentation
X = df.drop(['loan_id', 'loan_status'], axis=1)
y = df['loan_status']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Dataset
df.head(), X_train.shape, X_test.shape
```

# tune hyperparameters

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report

# RandomForestClassifier 모델 재정의
rf_model = RandomForestClassifier(random_state=42)

# 하이퍼파라미터 튜닝을 위한 GridSearchCV 설정
param_grid_rf = {
    'n_estimators': [50, 100, 150],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 4, 6],
    'min_samples_leaf': [1, 2, 3]
}

# GridSearchCV 객체 생성
grid_search_rf = GridSearchCV(estimator=rf_model, param_grid=param_grid_rf, cv=5, scoring='accuracy', verbose=1, n_jobs=-1)

# GridSearchCV를 사용하여 하이퍼파라미터 튜닝 및 교차 검증 수행
grid_search_rf.fit(X_train, y_train)

# 최적의 하이퍼파라미터와 그에 해당하는 정확도
best_params_rf = grid_search_rf.best_params_
best_accuracy_rf = grid_search_rf.best_score_

# 테스트 데이터셋에서 최적 모델의 성능 평가
best_rf_model = grid_search_rf.best_estimator_
y_pred_best_rf = best_rf_model.predict(X_test)
best_rf_accuracy_test = accuracy_score(y_test, y_pred_best_rf)
best_rf_report_test = classification_report(y_test, y_pred_best_rf)

best_params_rf, best_accuracy_rf, best_rf_accuracy_test, best_rf_report_test
```
